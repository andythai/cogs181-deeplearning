# Part 1 - Basic Matrix Operations Using Numpy

```python
# Part 1
import numpy as np 
A = np.array([[1,3],[5,7],[9,11]]) 
B = np.array([[1,-1],[-1,1],[-1,0]])
```

<b>(1) A − B</b>

```python
# Problem 1
print(A - B)
```

<b>(2) A ◦ B (element-wise product)</b>

```python
# Problem 2
print(A * B)
```

<b>(3) A<sup>T</sup>B</b>

```python
# Problem 3
print(np.dot(A.T, B))
```

<b>(4) AB<sup>T</sup></b>

```python
# Problem 4
print(np.dot(A, B.T))
```

<b>(5) AB</b>

```python
# Problem 5
try:
    np.dot(A, B)
except ValueError as e:
    print("Invalid operation: " + str(e))
    print("The operation AB is not valid: impossible")
```

# Part 2 - Basic Plots Using Matplotlib

```python
# Part 2
# Modified code to add 20 & 100 iterations
import matplotlib.pyplot as plt 
%matplotlib inline 
np.random.seed(0) 
space = np.linspace(0,10,num=50) 
sine = np.sin(space) 
plt.scatter(space,sine,color='b',label = 'sine_curve') 
sine_5 = sine 

# Add 20 and 100 iterative variables
sine_20 = sine
sine_100 = sine

for i in range(5): 
    sine_5 = sine_5 + np.random.normal(scale=0.1,size=50) 
    
# 20 iterations
for i in range(20): 
    sine_20 = sine_20 + np.random.normal(scale=0.1,size=50) 
# 100 iterations
for i in range(100): 
    sine_100 = sine_100 + np.random.normal(scale=0.1,size=50) 
    
plt_sine_5, = plt.plot(space,sine_5,color='r',label = 'noise_5_iters') 

# Plot 20 and 100 iteration graphs
plt_sine_20, = plt.plot(space,sine_20,color='g',label = 'noise_20_iters') 
plt_sine_100, = plt.plot(space,sine_100,color='purple',label = 'noise_100_iters') 

print('Solution graph:')
plt.legend(loc='upper right') 
plt.savefig('./cogs181_Q2.png')
```

# Part 6 - Linear regression

```python
# (1) Import matplotlib and numpy package into Python.
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# (2) Load data.
data = np.loadtxt('data.txt',dtype='float') 
x = data[:,0].reshape(len(data),1) 
y = data[:,1].reshape(len(data),1)
```

```python
# (3) Plot the data.
plt.plot(x,y) 
plt.grid()
plt.show()
```

```python
# (4) Create the matrix
X = np.hstack((np.ones((len(x),1)),np.power(x,1)))
```

```python
# (5) Compute the least squares line over the given data by
X_t = X.transpose((1,0)) 
sol = np.dot(np.linalg.inv(np.dot(X_t,X)),np.dot(X_t,y))
```

```python
# (6) Plot the least square line, and assign title to the ﬁgure
plt.plot(x,y) 
plt.hold(True) 
plt.plot(x,sol[0]+sol[1]*x) 
plt.title('Least square line fitting') 
plt.xlabel('x') 
plt.ylabel('y')
```
